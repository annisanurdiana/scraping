## Web Scraping with Pandas

> Extract All Table Elements From a Web Page Automatically

```python
import pandas as pd
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

web_scrapper = pd.read_html("https://en.wikipedia.org/wiki/Comparison_of_operating_systems")
web_scrapper
```

```python
for idx, table in enumerate(web_scrapper):
    print("----------------------------------------------------------------------")
    print(idx)
    print(table)
```
